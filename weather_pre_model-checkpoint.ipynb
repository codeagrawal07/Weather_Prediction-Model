```python
dataNum=dataset.select_dtypes(exclude='object')
dataNum.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,20))
re=1
for i in dataNum.columns:
    plt.subplot(3,9,re)
    sns.boxplot(dataset[i], palette='rainbow')
    plt.title(i)
    re+=1
plt.tight_layout()
```

```python
temQ1=dataset['visibility'].quantile(0.25)
temQ3=dataset['visibility'].quantile(0.75)
temIQR=temQ3-temQ1

visibility_lower=temQ1-(1.5*temIQR)
visibility_upper=temQ3+(1.5*temIQR)

winQ1=dataset['windspeed'].quantile(0.25)
winQ3=dataset['windspeed'].quantile(0.75)
winIQR=winQ3-winQ1

windspeed_lower=winQ1-(1.5*winIQR)
windspeed_upper=winQ3+(1.5*winIQR)


filtered_dataset = dataset[
    (dataset['visibility'] >= visibility_lower) & (dataset['visibility'] <= visibility_upper) &
    (dataset['windspeed'] >= windspeed_lower) & (dataset['windspeed'] <= windspeed_upper)
]
filtered_dataset.info()
```

```python
plt.figure(figsize=(20,20))
re=1
for i in dataNum.columns:
    plt.subplot(3,9,re)
    sns.boxplot(filtered_dataset[i], palette='rainbow')
    plt.title(i)
    re+=1
plt.tight_layout()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier  # Import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
#from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
# Prepare the data
X = filtered_dataset[['tempmax', 'tempmin', 'temp','feelslike','humidity','precipcover','windspeed','cloudcover','visibility']]  # Features
y = filtered_dataset['icon']  # Target variable

# Encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Create and train the decision tree model
model = DecisionTreeClassifier()  # Use DecisionTreeClassifier instead of LogisticRegression
model.fit(X_train, y_train)
#inputs=[72.1,55.3,61.5,61.5,56.8,0,5.4,37.6,15]
# Make predictions on the test set
y_pred = model.predict(X_test)

# Decode the predictions back to original labels
y_pred_decoded = label_encoder.inverse_transform(y_pred)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("score", model.score(X_test, y_test)*100)
DecisionTree_accuracy=model.score(X_test, y_test)*100
# Display predictions
#y_pred_decoded
# X_test['Predicted Icon'] = y_pred_decoded  # Adding predictions as a column
# X_test.to_csv('output_desionTree.csv', index=False)
```

```python
# Plot the decision tree
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree, DecisionTreeClassifier
plt.figure(figsize=(12, 8))
plot_tree(
    model,
    feature_names=X.columns,
    class_names=label_encoder.classes_,
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
svm=SVC()
svm.fit(X_train,y_train)
print('SVM accuracy :{: 2f}%'.format(svm.score(X_test,y_test)*100))
svm_accuracy=svm.score(X_test,y_test)*100
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
print('KNN accuracy :{: 2f}%'.format(knn.score(X_test, y_test)*100))
Knn_accuracy=knn.score(X_test,y_test)*100
```

```python
gbc=GradientBoostingClassifier(subsample=0.5,n_estimators=450,max_depth=5,max_leaf_nodes=25)
gbc.fit(X_train,y_train)
print('GBC accuracy :{: 2f}%'.format(gbc.score(X_test, y_test)*100))
gbc_accuracy=gbc.score(X_test,y_test)*100
```

```python

accuracies = {
    'Decision Tree':DecisionTree_accuracy, 
    'KNN': Knn_accuracy,
    'SVM': svm_accuracy,
    'GBC': gbc_accuracy
}
models = list(accuracies.keys())
scores = list(accuracies.values())

# Create the bar chart
plt.figure(figsize=(5, 6))
plt.bar(models, scores, color=['blue', 'green', 'orange', 'purple'])

# Add labels and title
plt.xlabel('Models', fontsize=14)
plt.ylabel('Accuracy (%)', fontsize=14)
plt.title('Comparison of Model Accuracies', fontsize=11)
plt.ylim(0, 100)  # Set the y-axis range to 0-100 for percentage
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate bars with their accuracy scores
for i, score in enumerate(scores):
    plt.text(i, score + 1, f'{score:.1f}%', ha='center', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
